### 와인 가격대 시각화
 - 어떤 그래프?
     - 미세한 분포를 가진 가격들의 막대 그래프를 그리기에는 적합하지 않을 것이라 판단
     - 가격 변화 추이가 아닌 가격대를 확인하고 싶기 때문에 곡선 그래프도 적합하지 않을 것
     
 - Box와 Whisker를 통해 중앙값 등을 확인하기 용이한 Box Plot을 선정!

```python
# 범위 설정을 위해 최대값 확인
df['WINE_PC'].max()
```

```python
# 그래프 스케일 max 값을 10000으로 잡지 않고 200 정도로 잡아 모양을 보자!

plt.style.use('default')
plt.rcParams['figure.figsize'] = (4, 6)
plt.rcParams['font.size'] = 12

price = df['WINE_PC']

fig, ax = plt.subplots()
box = ax.boxplot([price], notch=True, whis=2.5)
ax.set_ylim(0, 200)
ax.set_xlabel('WINE')
ax.set_ylabel('PRICE')
```

```python
whiskers = [item.get_ydata() for item in box['whiskers']]
medians = [item.get_ydata() for item in box['medians']]

print('whiskers:', whiskers)  #수염 부분 값
print('medians:', medians)  #중앙값
```

### 와인 종류 별 분포
 - 데이터 안에 와인이 종류 별로 몇 개씩 들어가있는지 그 개수를 확인해보자!
 - 종류 별 개수의 차이를 확인해보기 위해 막대 그래프를 그려보자!

```python
category = df['WINE_CAT'].unique().tolist()
category
```

```python
# 리스트 속 nan값 제거
category = [x for x in category if not(pd.isnull(x)) == True]
```

```python
category
```

```python
cat_num = df['WINE_CAT'].value_counts()
cat_num
```

```python
index = np.arange(len(category))

plt.figure(figsize=(10,3))
plt.bar(index, cat_num)
plt.title('Number of Wines per Category', fontsize=20)
plt.xlabel('Categories', fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.xticks(index, category, fontsize=8, rotation=90)
plt.show()
```

### 와인 제조년도 별 개수
 - pandas를 활용해 와인의 제조년도를 뽑아내고 제조년도 별로 몇 개씩 들어가있는지 그 개수를 확인해보자!


```python
df['year'] = df['WINE_NM'].str.split(' ').str[-1]
```

```python
df['year']
```

```python
df1 = df.copy()
```

```python
df1['year'].unique()
```

```python
df1['year'].replace(r"[\D]", np.nan, regex=True, inplace=True)  # [\D]: 숫자가 아닌 값 제거
```

```python
df1['year'].unique()
```

```python
df1[df1['year'].isna() == True]
```

```python
nan_list = df1[df1['year'].isna() == True].index.tolist()
nan_list
```

```python
df1.drop(labels=nan_list, inplace=True)
```

```python
df1['year'].unique()
```

```python
df1[~(df1['year'].astype('int32') > 1000)]
```

```python
# 연도가 이상한 행의 index 추출 후 list에 넣음
odd_idx = df1.index[~(df1['year'].astype('int32') > 1000)].tolist()
odd_idx
```

```python
df1.drop(labels=odd_idx, inplace=True)
```

```python
year = df1['year'].unique().tolist()
year
```

```python
year_num = df1['year'].value_counts()
year_num
```

```python
index = np.arange(len(year))

plt.figure(figsize=(10,3))
plt.bar(index, year_num, color='r')
plt.title('Wine Vintages', fontsize=20)
plt.xlabel('Year', fontsize=18)
plt.ylabel('Count', fontsize=18)
plt.xticks(index, year, fontsize=8, rotation=90)
plt.show()
```

### 와인의 지역 별 개수와 그에 따른 가격
 - 데이터 내의 와인들이 어떤 지역에 주로 분포되어 있는지 살펴보고 가격이 어떻게 형성되어 있는지 살펴보자!

```python
df['region'] = df['WINE_REGION_NM'].str.split(' ').str[-1]
```

```python
df['region']
```

```python
df2 = df.copy()
```

```python
df2['region'].unique()
```

```python
df2[df2['region'].str.contains("Zealand")].sample(20)
```

```python
df2['region'] = df2['region'].str.replace("Zealand", "New Zealand")
```

```python
df2['region'] = df2['region'].str.replace("\(FYROM\)", "Fyrom")
```

```python
df2[df2['region'].str.contains("York")]
```

```python
df2['region'] = df2['region'].str.replace("York", "New York")
```

```python
df2[df2['region'].str.contains("from")]
```

```python
df2.drop(df2[df2['region'].str.endswith("from")].index, inplace = True)
```

```python
df2['region'].unique()
```

```python
df3 = df1.join(df2['region'])
df3
```

```python
region_list = df3['region'].unique().tolist()
region_list = [x for x in region_list if not(pd.isnull(x)) == True]
region_list = sorted(region_list)
region_list
```

```python
df4 = df3[['region','WINE_PC']].sort_values(by='region').dropna()
df4
```

```python
price_mean = df4.groupby('region')['WINE_PC'].mean().tolist()
price_mean
```

```python
plt.figure(figsize=(15,8))
plt.bar(region_list, price_mean)
plt.xticks(region_list, fontsize=10, rotation=90)
plt.show()
```
